```python
%matplotlib inline
import numpy as np
```

```python
import matplotlib.pylab as plt
```

This is a test jupyter notebook 

the notebook shows how to produce a basic plot in jupyter environment. Starting with defining 3 frequencies

```python
freqs=[1,2,3]
nfreq=len(freqs)
```

The plot will be shown in the range from 0 to 1!

```python
x=np.arange(0,1,0.01)
```

Let us define an empty array that will afterwards be filled up with a for-loop.

```python
y=np.zeros((nfreq,len(x)))
```

```python
for i in range(nfreq):
    y[i,:] = np.sin(freqs[i]*2*np.pi*x)
```

Finally, the three different frequnecies are shall be shown in one plot

```python
plt.figure()
plt.plot(x,y.T)
plt.show()
```
